```python
import os
import torch
import logging
import numpy as np
from datetime import datetime
from typing import Optional

from dataclasses import dataclass, field
from transformers import HfArgumentParser

import matplotlib.pyplot as plt
```

```python
results = torch.load('results/structured-results.pt')
```

```python
results.keys()
```

```python
x = range(1, len(results['f1']))
fig, ax = plt.subplots()

ax.plot(x, results['precision'][:-1], label="precision")
ax.plot(x, results['recall'][:-1], label='recall')
ax.plot(x, results['f1'][:-1], label="f1")
ax.plot(x, results['accuracy'][:-1], label="accuracy")
ax.set_title("Evaluation performance w/ cross validation")
ax.set_xlabel('# epoch')
ax.set_ylabel('Metric scores')
ax.legend()

f_name = f'sup-eval.png'
fig.savefig(os.path.join('../../figures', f_name), dpi=150, bbox_inches='tight')
```
